```python
target_dict = {}
for k in fs.keys():
    target_dict[k] = fs[k].loc[['PER(배)', 'ROE(지배주주)', '부채비율', '영업이익률', '배당성향(%)']]
```

```python
target_dict
```

```python
dic = {}
for k in target_dict.keys():
    ls = []
    ls += (list(target_dict[k].loc['PER(배)']))
    ls += (list(target_dict[k].loc['ROE(지배주주)']))
    ls += (list(target_dict[k].loc['부채비율']))
    ls += (list(target_dict[k].loc['영업이익률']))
    ls += (list(target_dict[k].loc['배당성향(%)']))
    dic[k] = ls
```

```python
df = pd.DataFrame(dic).T
```

```python
t = ['per', 'roe', 'dr', 'opm', 'dpr']
c = []
for i in t:
    for r in range(1,7):
        c.append(i + str(r))
```

```python
df.columns = c
```

```python
df = df.reset_index()
```

```python
col = df.columns[1:]
for c in col:
    df[c] = df[c].replace('-', np.nan)
    df[c] = df[c].astype(float)
```

```python
df.to_csv('target_fs.csv', index=False)
```

```python
from tqdm import tqdm

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import yfinance as yf
spy = yf.Ticker("SPY").history(period="max")
spy.index = spy.index.map(lambda x: x.date())
spy = spy[pd.to_datetime(spy.index) >= "2020-05-01"]
return_spy = spy["Close"].pct_change()
buying_day = [True if i.day == 15 else False for i in return_spy.index]
action = return_spy * np.nan
action.loc[buying_day] = 1  # 한 달에 한 번씩 사는 것을 가정
action.loc[buying_day] = action.dropna().cumsum()
action = action.ffill()  # 매달 한 주식씩 더 사는 것을 가정
(1 + return_spy.fillna(0)).cumprod().plot(figsize=(10, 3), grid=True)
plt.show()

(1 + (return_spy * action.shift(1)).fillna(0)).cumprod().plot(
    figsize=(10, 3), grid=True
)
plt.show()
```
